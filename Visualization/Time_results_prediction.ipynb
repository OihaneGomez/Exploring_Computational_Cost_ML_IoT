# Time results

The prediction process is tested to evaluate how long it takes to read and process a sequence of data and compare it with the already trained models. The experiment consists of measuring the elapsed time in the process of applying every machine learning technique on a sample of 14 instances of the data set.


### Data Processing Times Vs Number of features

* XYZ and the most representative component (X) for Laptop, RPi and RZero

Computational cost of reading and processing the 14 instances of new data for a variable number of features

* Barplot prediction processing times 

Comparative of the processing times for all the features, top 10 and top 3

### Inference task times Vs Number of features

Computational cost of inference task according to every algorithm when data is already processed.

* XYZ and the most representative component for Laptop, RPi and RZero

* LG, NV, SVM, MLP, DT algorithms (KNN and RF are out of this representation since they have a significantly greater duration than those included)

```python
#txt file containing time values
time_file_pred_laptop = 'txt_files/Predicion_task_Laptop.txt'
time_file_pred_laptop_std = 'txt_files/Predicion_task_Laptop_std.txt'

time_file_pred_Rpi = 'txt_files/Predicion_task_Rpi.txt'
time_file_pred_Rpi_std = 'txt_files/Predicion_task_Rpi_std.txt'

time_file_pred_RZero = 'txt_files/Predicion_task_RZero.txt'
time_file_pred_RZero_std = 'txt_files/Predicion_task_RZero_std.txt'
```
