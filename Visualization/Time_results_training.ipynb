# Time results

The training process can be divided into two parts, the common pre-processing steps applied to all the data set and the algorithm-dependent model creation task.


### Data Processing Times Vs Number of features

* XYZ and the most representative component (X) for Laptop, RPi and RZero

Elapsed time for the initial pre-processing stage when all the data set information is read, and the features are calculated

* Barplot prediction processing times 

Comparative of the processing times for all the features, top 10 and top 3

### Fitting task times Vs Number of features

Average computational cost of every algorithm for the model fitting task. The second step of the training stage is feed the model with the pre-processed data for the fitting task. This task depends on the different classifiers and their parameters, together with the amount of data used to build it. 

* XYZ and the most representative component for Laptop, RPi and RZero

* LG, NV, SVM, KNN, DT algorithms (RF and MLP are out of this representation since they have a significantly greater duration than those included)

```python
#txt file containing time values
time_file_fit_laptop = 'txt_files/Fitting_task_Laptop.txt'
time_file_fit_laptop_std = 'txt_files/Fitting_task_Laptop_std.txt'

time_file_fit_Rpi = 'txt_files/Fitting_task_Rpi.txt'
time_file_fit_Rpi_std = 'txt_files/Fitting_task_Rpi_std.txt'

time_file_fit_RZero = 'txt_files/Fitting_task_RZero.txt'
time_file_fit_RZero_std = 'txt_files/Fitting_task_RZero_std.txt'
```
